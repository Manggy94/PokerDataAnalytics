```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from src.data.loader import DataLoader

data_loader = DataLoader()
data = data_loader.fast_load_showdown_hands()
print(data.info())
data.head()
```

```python
from src.pipelines.preprocessing.models.random_forest import RandomForestFeaturesPreprocessor
preprocessor = RandomForestFeaturesPreprocessor()
features = preprocessor.fit_transform(data)
print(features.info())
features.head()
```

```python
targets = data['player_combo_hand']
print(targets.info())
targets.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, top_k_accuracy_score
```

```python
print(accuracy_score(y_test, y_pred))
top_k_accuracy_score(y_test, y_pred, k=5)
top_k_accuracy_score(y_test, y_pred, k=10)
top_k_accuracy_score(y_test, y_pred, k=15)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
